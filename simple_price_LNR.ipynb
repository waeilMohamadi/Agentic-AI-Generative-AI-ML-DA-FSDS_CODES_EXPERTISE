```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Load Your DataSet

```python
dataset = pd.read_csv(r"C:\Users\waeil\OneDrive\Desktop\ML_AI\15th, 16th - SLR\SLR - House price prediction\House_data.csv")
```

```python
dataset.head(4)
```

# Divide the dataset int Feature(X) and Target(Y)

```python
space=dataset['sqft_living'] # independent variable (X)
price=dataset['price']      # dependent variable (Y)

x = np.array(space).reshape(-1, 1) # reshaping to make it a 2D array
y = np.array(price) # dependent variavle as 1D array
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=1/3, random_state=0) # splitting the dataset into training and testing sets

#Fitting simple linear regression to the Training Set
model = LinearRegression() # create an object of model
model.fit(x_train, y_train) # fit the model to the training data

#Predicting the prices
y_pred = model.predict(x_test) # predicting the prices for test data


#Visualizing the training Test Results 
plt.scatter(x_train, y_train, color= 'red')
plt.plot(x_train, model.predict(x_train), color = 'blue')
plt.title ("Visuals for Training Dataset")
plt.xlabel("Space")
plt.ylabel("Price")
plt.ticklabel_format(style='plain', axis='y') # shows full number instead of scientific notation
plt.show()


#Visualizing the Test Results 
plt.scatter(x_test, y_test, color= 'red')
plt.plot(x_train, model.predict(x_train), color = 'blue')
plt.title("Visuals for Test DataSet")
plt.xlabel("Space")
plt.ylabel("Price")
plt.ticklabel_format(style='plain', axis='y') # shows full number instead of scientific notation
plt.show()
```
