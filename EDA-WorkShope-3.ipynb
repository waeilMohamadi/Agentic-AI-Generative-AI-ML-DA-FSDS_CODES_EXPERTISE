# Handle Remaining Missing Values
- Age has NaN in rows 2 and 3

- Location has NaN in rows 2 and 4

- Exp has NaN in row 3

```python
emp
```

```python
# Fill numeric columns with mean
#emp['Age'] = emp['Age'].fillna(emp['Age'].mean())
#emp['Exp'] = emp['Exp'].fillna(emp['Exp'].mean())
```

```python
# Assuming your dataset is already cleaned
# Convert numeric columns
emp['Age'] = pd.to_numeric(emp['Age'], errors='coerce') #errors='coerce' ensures that any leftover strings turn into NaN
emp['Exp'] = pd.to_numeric(emp['Exp'], errors='coerce')
emp['Salary'] = pd.to_numeric(emp['Salary'], errors='coerce')

# Fill missing values
emp['Age'] = emp['Age'].fillna(emp['Age'].mean())
emp['Exp'] = emp['Exp'].fillna(emp['Exp'].mean())
emp['Location'] = emp['Location'].fillna(emp['Location'].mode()[0])
```
