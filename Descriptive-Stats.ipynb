```python
# First Way
# This will produce a table where each unique value of Mthly_HH_Expense is counted.
monthly_exp = pd.crosstab(index=income_Exp["Mthly_HH_Expense"],columns="count")
monthly_exp.reset_index(inplace=True)
monthly_exp[monthly_exp['count'] == income_Exp.Mthly_HH_Expense.value_counts().max()] # gives the highest frequency(mode)
```

```python
# Second Way
monthly_exp = income_Exp["Mthly_HH_Expense"].value_counts().reset_index(name="High Income")
monthly_exp.columns = ["Mthly_HH_Expense", "count"]
monthly_exp[monthly_exp['count'] == monthly_exp['count'].max()]
```

```python
# Third Way
income_Exp["Mthly_HH_Expense"].mode()[0] # [0] Selects the first value from the mode Series (useful if there are multiple modes).
```

```python
# fourth Way
counts = income_Exp["Mthly_HH_Expense"].value_counts()
counts.idxmax(), counts.max()
```
