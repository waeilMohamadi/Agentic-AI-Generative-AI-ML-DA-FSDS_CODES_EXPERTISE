```python
# Load image
image = Image.open(r'C:\Users\THINKPAD E15\Desktop\w.jpg')

# Show original image
plt.imshow(image)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python
gray = image.convert('L')  # 'L' = 8-bit grayscale
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
# Blur
blurred = image.filter(ImageFilter.BLUR)
```

```python
blurred
```

```python
# Sharpen
sharpened = image.filter(ImageFilter.SHARPEN)
sharpened
```

```python
# Edge detection
edges = image.filter(ImageFilter.FIND_EDGES)
edges
```

```python
# Display all
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(blurred); ax[0].set_title("Blurred")
ax[1].imshow(sharpened); ax[1].set_title("Sharpened")
ax[2].imshow(edges); ax[2].set_title("Edges")
for a in ax: a.axis('off')
plt.show()
```
