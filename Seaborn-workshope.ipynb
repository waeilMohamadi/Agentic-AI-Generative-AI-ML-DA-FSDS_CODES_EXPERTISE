# Introduction to Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pn
plt.rcParams["figure.figsize"]= 10,3
exercies = sns.load_dataset("exercise")
g = sns.catplot(x = "time", y="pulse", hue="kind", data=exercies)
plt.grid()
plt.show()
```

```python
plt.rcParams["figure.figsize"]= 10,3
exercies = sns.load_dataset("exercise")
g = sns.violinplot(x = "time", y="pulse", hue="kind", data=exercies)
plt.show()
```

```python
plt.rcParams["figure.figsize"]= 10,3
exercies = sns.load_dataset("exercise")
g = sns.stripplot(x = "time", y="pulse", hue="kind", data=exercies)
plt.show()
```

```python
plt.rcParams["figure.figsize"]= 10,3
exercies = sns.load_dataset("exercise")
g = sns.boxplot(x = "time", y="pulse", hue="kind", data=exercies)
plt.show()
```

```python
sns.set(style = 'ticks') # How to change the color
tip = sns.load_dataset("tips")
load = sns.stripplot(x="day",y="total_bill", data = tip)
plt.show()
```

```python
sns.set(style = 'darkgrid') # How to change the color
tip = sns.load_dataset("tips")
load = sns.stripplot(x="day",y="total_bill", data = tip)
plt.show()
```
