# FORMAL ARGUMENT & ACTUAL ARGUMENT

```python
def student(name,age): # formal argument
    print(name)
    print(age)
student('Fahd',3,4) # actual argument
```

```python
def student(name,age): # formal argument
    print(name)
    print(age)
student(2,'Fahd') # actual argument
```

```python
def student(name,age=3): # default argument
    print(name)
    print(age)
student(2) # actual argument
```

# Default arguments in Python are parameters that have a default value if no value is provided by the caller.

```python
def student(name,age=3): # default argument
    print(name)
    print(age)
student('Fahd',5) # actual argument
```

# keyword arguments
- Keyword arguments in Python are arguments you pass to a function using the parameter name, like this:
- student(name='Fahd', age=2)

```python
def student(name,age): # keyword argument
    print(name)
    print(age)
student(name='Fahd',age=2) # actual argument
```

```python
def student(name,age): # keyword argument
    print(name)
    print(age)
student(age=2,name='Fahd') # actual argument
```

# variable-length argument
- *ages collects any number of positional arguments into a tuple.

```python
def student(name,*age): # variable length argument
    print(name)
    for i in age:
        print(i)
student('Fahd',3,2,1,5) # actual argument
```

# keyword arguments for **age

```python
def student(name, **age):
    print(name)
    print(age)

student(name='Fahd', age1=5, age2=6, age3=7, age4=8, age5=9)
```

```python
# in progammin we need to continue these process thats why we need to use loop hear
def fib(n):
    a = 0
    b = 1
    print(a)
    print(b)
    for i in range(0, n):
        c = a + b
        a = b
        b = c
        print(c)  
fib(5)
```

# Factorial

```python
def fact(n):
    fatctorial = 1
    for i in range(1,n-1):
        fatctorial = fatctorial*i
    return fatctorial
print(fact(10))
```

```python
def fact(n):
    factorial = 1
    for i in range(1, n + 1):
        factorial = factorial * i
    return factorial

print(fact(5))  # Output: 120
```
