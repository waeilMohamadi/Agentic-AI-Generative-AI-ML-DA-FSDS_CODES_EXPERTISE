```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,6))
```

```python
plt.scatter(stats["InternetUsers"], stats["BirthRate"], c='blue', alpha=0.6)

plt.title("Birth Rate vs Internet Users by Country")
plt.xlabel("Internet Users (%)")
plt.ylabel("Birth Rate (per 1000 people)")
plt.grid(True)
plt.show()
```
